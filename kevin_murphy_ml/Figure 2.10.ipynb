```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import scipy.stats
```

```python
params = [(0.1, 0.1), (1., 1.), (2., 3.), (8., 4.)]

x = np.linspace(0, 1, 100)

fig, ax = plt.subplots()
for a, b in params:
    rv = scipy.stats.beta(a, b)
    ab_val_str ='a=%s,b=%s' % (a, b)
    print ab_val_str
    print rv.mean(), a/(a+b), rv.rvs(1000).mean()
    print rv.var(), a*b / ((a+b)**2 * (a+b+1)), rv.rvs(1000).var()
    ax.plot(x, rv.pdf(x), label=ab_val_str)
    ax.legend()
```
